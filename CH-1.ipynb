# Data Cleaning

remove null values

```python
import pandas as pd
df=pd.read_csv('sales_data.csv')
print(df)
```

```python
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
df.dropna()#inplace=True
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='any')
```

```python
df.dropna(subset=['sales','expenses'])
```

```python
df['sales'].fillna(12)
df['sales'].fillna(df['sales'].mean())
```

```python
df.fillna(1)
```
